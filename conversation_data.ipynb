```python
conversation_data = {
    "hello":"Hi there! How can I help you today",
    "hi":"Hello! How are you",
    "how are you":"I'm just bot ,but I'm doing great! How about you?",
    "what is computer vision":"Computer vision is a field of AI that allows computer to see",
    "what is ai":"AI stands for Artificial Intelligence,which anabels machines to perform task",
    "what is deep learning":"Deep Learning is a subset of AI that uses neural networks to learn",
    "bye":"Goodbye! Have a nice day!",
    "thank you":"you're welcome! Happy to help"
}
```

```python
def simple_chat(user_input, chat_history=[]):
  user_input_lower = user_input.lower()
  response = None

  for key in conversation_data:
    if key in user_input_lower:
      response = conversation_data[key]
      break

  if not response:
    response = "Sorry,I don't understand that.can you ask something else?"

  chat_history.append(("you", user_input))
  chat_history.append(("Bot", response))

  return response, chat_history
```

```python
print("Welcome to Mini Conversational AI")
print("Type 'exit' or 'quit' to end the chat.\n")

chat_history =[]

while True:
  user_input = input("you:")

  if user_input.lower() in ["exit","quit"]:
    print("chat ended.Goodbye!")
    break

  bot_response,chat_history = simple_chat(user_input,chat_history)
  print("Bot:", bot_response)
```

```python
print("\n===== Full Chat History=====")
for speaker,message in chat_history:
  print(f"{speaker}: {message} ")

  # useful for debugging or showing what happened in the session.
```
